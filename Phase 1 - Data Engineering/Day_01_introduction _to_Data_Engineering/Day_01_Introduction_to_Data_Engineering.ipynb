### 1. Dataset Overview

```python
# Total number of students
total_students = df.shape[0]
print(f"Total number of students: {total_students}")

# Total number of columns
total_columns = df.shape[1]
print(f"Total number of columns: {total_columns}")

# Number of unique departments
unique_departments = df['department'].nunique()
print(f"Number of unique departments: {unique_departments}")
```

### 2. Department Wise Analysis

```python
department_counts = df['department'].value_counts()
print("Count of students in each department:")
print(department_counts)
```

### 3. Score Analysis

```python
# Highest math score
highest_math = df['math_score'].max()
print(f"Highest math score: {highest_math}")

# Lowest math score
lowest_math = df['math_score'].min()
print(f"Lowest math score: {lowest_math}")

# Average math score
avg_math = df['math_score'].mean()
print(f"Average math score: {avg_math:.2f}")

# Average programming score
avg_programming = df['programming_score'].mean()
print(f"Average programming score: {avg_programming:.2f}")

# Average attendance percentage
avg_attendance = df['attendance_percentage'].mean()
print(f"Average attendance percentage: {avg_attendance:.2f}%")
```

### 4. Gender Wise Analysis

```python
avg_total_score_by_gender = df.groupby('gender')['total_score'].mean()
print("Average total score by gender:")
print(avg_total_score_by_gender)
```

### 5. Top Performers

```python
top_5_students = df.sort_values(by='total_score', ascending=False).head(5)
print("Top 5 students based on total score:")
print(top_5_students[['name', 'department', 'total_score']])
```

### 6. Attendance Analysis

```python
low_attendance_students = df[df['attendance_percentage'] < 75]

if not low_attendance_students.empty:
    print("Students with attendance below 75%:")
    print(low_attendance_students[['name', 'department', 'attendance_percentage']])
else:
    print("All students have attendance above 75%.")
```

###1.How many students are from the Electronics Department?

```python
electronics_count = len(df[df["department"] == "ECE"])
print("Students from Electronics Department :", electronics_count)
```

###2.What is the average attendance of female students?

```python
female_attendance = df[df["gender"] == "Female"]["attendance_percentage"].mean()

print("Average attendance of female students :", female_attendance)
```

###3. Who is the student with the lowest programming score?

```python
lowest_programming = df.loc[df["programming_score"].idxmin()]

print("Student with lowest programming score :", lowest_programming["name"])
```

###4. How many students scored above 80 in math?

```python
above_80_math = len(df[df["math_score"] > 80])

print("Students scored above 80 in math :", above_80_math)
```

###5. What is the average total_score for students with attendance above 90%?

```python
avg_total_above_90 = df[df["attendance_percentage"] > 90]["total_score"].mean()

print("Average total score :", avg_total_above_90)
```
